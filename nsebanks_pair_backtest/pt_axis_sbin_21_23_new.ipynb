```python
df.set_index('date', inplace=True)

# Resample the dataframe to daily data
df_resampled = df.resample('D').last()

# Display the resampled dataframe
print(df_resampled)
```

```python
df_resampled.dropna(inplace=True)
```

```python
df_resampled.reset_index(inplace=True)
```

```python
df_resampled['Date']=df_resampled['date'].dt.date
```

```python
df_resampled.drop('date',axis=1,inplace=True)
```

```python
df_resampled['Date'] = pd.to_datetime(df_resampled['Date'])

# Set 'Date' column as index
df_resampled = df_resampled.set_index('Date')
```

```python

# Assuming df_resampled is your dataframe and 'Date' is your index column and 'ratio' is your column for calculation
df_resampled = df_resampled.sort_index()  # making sure the dataframe is sorted by date

# First set the initial start and end dates
start_date = pd.Timestamp('2019-01-01')
end_date = pd.Timestamp('2020-12-31')

# Create an empty list to store the calculated means
means = []
stdevs=[]
# Calculate means for the specified period and keep shifting the window by one day
while end_date <= df_resampled.index[-1]:
    mean_ratio = df_resampled.loc[start_date:end_date, 'ratio'].mean()  # calculate the mean
    stdev_x=df_resampled.loc[start_date:end_date, 'ratio'].std()
    means.append(mean_ratio)  # append the mean to our list
    stdevs.append(stdev_x)
    # Shift the window by one day
    start_date += pd.DateOffset(days=1)
    end_date += pd.DateOffset(days=1)

# Now create a new dataframe with dates and means, then merge it back into the original dataframe
df_means = pd.DataFrame(means, index=pd.date_range(start='2019-01-01', periods=len(means), freq='D'), columns=['mean_ratio'])
df_stdevs=pd.DataFrame(stdevs, index=pd.date_range(start='2019-01-01', periods=len(stdevs), freq='D'), columns=['stdev_x'])

# Merge the new dataframe with the original dataframe
df_resampled = df_resampled.merge(df_means, left_index=True, right_index=True, how='left')
df_resampled = df_resampled.merge(df_stdevs, left_index=True, right_index=True, how='left')
```

```python
df_resampled.reset_index(inplace=True)
```
