```python
# get average daily return times 252 trading days a year
e_r = df.mean()*252

# plot expected return
e_r.plot(kind='bar')
plt.title('Expected Annual Return');
```

```python
sd = df.std()*np.sqrt(252) 
sd.plot(kind='bar') 
plt.title('Standard Deviation (Volatility)');
```

```python
import seaborn as sns
```

```python
cov_matrix = df.cov()
cov_matrix
sns.heatmap(df.corr(),annot=True);
```

```python
p_ret = []
p_weights = []
p_vol = []

#set no. of assets and portfolios to simulate
num_assets = len(df.columns)
num_portfolios = 10_000_000

#for each random portfolio, find the return and volatility
for portfolio in range(num_portfolios):
    #create random weights
    weights = np.random.random(num_assets)
    weights = weights/np.sum(weights)
    p_weights.append(weights)
    
    returns = np.dot(weights,e_r)
    p_ret.append(returns)
    
    var = cov_matrix.mul(weights,axis=0).mul(weights,axis=1).sum().sum()
    ann_sd = np.sqrt(var*252)
    p_vol.append(ann_sd)
    
data = {'returns': p_ret, 'volatility': p_vol}

for counter, symbol in enumerate(df.columns.tolist()):
    data[symbol + '_weight'] = [w[counter] for w in p_weights]
    
portfolios = pd.DataFrame(data)

portfolios.to_csv('/Users/divyansh/Documents/ACA/efficient_frontier_mpt/portfolios.csv',index=False)
```

```python
#portfolios with lowest volatility
min_var_port = portfolios.loc[portfolios['volatility'].idxmin()]
plt.figure(figsize=[12, 8])
plt.scatter(x=portfolios['volatility'], y=portfolios['returns'], c=portfolios['sharpe'], cmap='viridis', marker='o', s=10, alpha=0.3)
plt.xlabel('Volatility')
plt.ylabel('Returns')
plt.colorbar(label='Sharpe Ratio')

#portfolios.plot.scatter(x='volatility', y='returns', c='sharpe', cmap='viridis', marker='o', s=10, alpha=0.3,figsize=[10,10])
plt.scatter(x=min_var_port[1],y=min_var_port[0],\
            color='r', marker='*', s=500)
plt.title('Efficient Frontier and global minimum variance portfolio')
plt.show()
```
