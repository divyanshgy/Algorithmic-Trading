```python
import numpy as np 
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
df_1=pd.read_csv('/Users/divyansh/Documents/ACA/NIfty_50_Stocks_min_data_17_23/HCLTECH_clean.csv')
df_2=pd.read_csv('/Users/divyansh/Documents/ACA/NIfty_50_Stocks_min_data_17_23/TECHM.csv')
```

```python
df_1['date'] = pd.to_datetime(df_1['date'])
start_date = pd.to_datetime('2017-01-02 09:15:00+05:30')
end_date = pd.to_datetime('2019-12-31 15:29:00+05:30')
df_s1 = df_1[(df_1['date'] >= start_date) & (df_1['date'] <= end_date)]
```

```python
df_s1
```

```python
df_s1.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df_s1.drop('volume',axis=1,inplace=True)
```

```python
df_s1
```

```python
df_s1['Date']=df_s1['date'].dt.date
df_s1['Time']=df_s1['date'].dt.time
df_s1.drop(['date'],axis=1,inplace=True)
```

```python
df_s1
```

```python
df_2['date'] = pd.to_datetime(df_2['date'])
start_date = pd.to_datetime('2017-01-02 09:15:00+05:30')
end_date = pd.to_datetime('2019-12-31 15:29:00+05:30')
df_s2 = df_2[(df_2['date'] >= start_date) & (df_2['date'] <= end_date)]
```

```python
df_s2
```

```python
df_s2.drop('Unnamed: 0',axis=1,inplace=True)
df_s2.drop('volume',axis=1,inplace=True)
```

```python
df_s2
```

```python
df_s2['Date']=df_s2['date'].dt.date
df_s2['Time']=df_s2['date'].dt.time
df_s2.drop(['date'],axis=1,inplace=True)
```

```python
df_s2
```

```python
df=pd.DataFrame()
df['ratio']=df_s1['close']/df_s2['close']
df['Date']=df_s1['Date']
df['Time']=df_s1['Time']
```

```python
df
```

```python
df.isnull().sum()
```

```python
mean=np.mean(df['ratio'])
stdev=np.std(df['ratio'])
```

```python
print(mean)
print(stdev)
```

```python
from datetime import timedelta
```

```python
df.plot(x='Date',y='ratio',kind='line',label='Spread Ratio')
plt.axhline(mean, color='black', linestyle='--', label='Mean')
plt.axhline(mean - stdev, color='red', linestyle='--', label='-1 Std Dev')
plt.axhline(mean - (2*stdev), color='red', linestyle='--', label='-2 Std Dev')
plt.axhline(mean - (1.5*stdev), color='red', linestyle='--', label='-2 Std Dev')
plt.axhline(mean + stdev, color='green', linestyle='--', label='+1 Std Dev')
plt.axhline(mean + (2*stdev), color='green', linestyle='--', label='+2 Std Dev')
plt.axhline(mean + (1.5*stdev), color='green', linestyle='--', label='+1 Std Dev')
plt.title('Spread Ratio')
plt.xticks(rotation=45)
plt.show()
```

```python
print(mean+(2.75*stdev))
```

```python
results=[]
total_pnl=0
position_1=0
sl=0
signal=None
trade_status=None
entry_1=None
entry_2=None
entry_date=None
entry_time=None
exit_1=None
exit_2=None
exit_date=None
exit_time=None
pnl=None
# entry_date=None
# entry_time=None

dates=df['Date'].unique()

for date in dates:
    Date=date
    daily_date=df[df['Date']==date]
    daily_date_s1=df_s1[df_s1['Date']==date]
    daily_date_s2=df_s2[df_s2['Date']==date]
    
    time_range=daily_date[(daily_date['Time']>pd.to_datetime('9:15:00').time()) & (daily_date['Time']<pd.to_datetime('15:30:00').time())]['Time'].unique()
    
    for time in time_range:
        if (position_1 ==0) and (daily_date[daily_date['Time']==time]['ratio'].iloc[0]>=(mean+(2*stdev))):
            position_1=1
            signal='Sell'
            entry_1=daily_date_s1[daily_date_s1['Time']==time]['close'].iloc[0]
            entry_2=daily_date_s2[daily_date_s2['Time']==time]['close'].iloc[0]
            entry_date=date
            entry_time=time
            
        elif (position_1==1):
            sl=mean+(2.75*stdev)
            if (daily_date[daily_date['Time']==time]['ratio'].iloc[0]<=(mean)):
                position_1==0
                trade_status='Trade complete'
                exit_1=daily_date_s1[daily_date_s1['Time']==time]['close'].iloc[0]
                exit_2=daily_date_s2[daily_date_s2['Time']==time]['close'].iloc[0]
                exit_date=date
                exit_time=time
                pnl=(entry_1-exit_1)+(exit_2-entry_2)
                
            elif (daily_date[daily_date['Time']==time]['ratio'].iloc[0]>=sl):  
                position_1==0
                trade_status='SL hit'
                exit_1=daily_date_s1[daily_date_s1['Time']==time]['close'].iloc[0]
                exit_2=daily_date_s2[daily_date_s2['Time']==time]['close'].iloc[0]
                exit_date=date
                exit_time=time
                pnl=(entry_1-exit_1)+(exit_2-entry_2)
                
            #total_pnl+=pnl
            
        #print(total_pnl)
                
                
    result = {
        'Date': date,
        'signal' : signal,
        'Trade Status': trade_status,
        'Entry Price s1': entry_1,
        'Entry Price s2': entry_2,
        'Entry Date': entry_date,
        'Entry time': entry_time,
        'Exit Price s1': exit_1,
        'Exit Price s2': exit_2,
        'Exit Date': exit_date,
        'Exit Time': exit_time,
        'PnL': pnl,
    }     

    results.append(result)
results_df=pd.DataFrame(results)
print(results_df)
```
